subplot and legend

```python
days=[1, 2, 3, 4, 5]
original_price=[10, 20, 25, 30, 40]
predicted_price=[8, 21, 25, 28, 38]
```

```python
plt.plot(days, original_price, 'b-o' ,label='Original price')
plt.plot(days, predicted_price, 'r-o',label='Predicted price')
plt.legend()
plt.title('Plot with legends')
plt.xlabel('Weekdays')
plt.ylabel('price')
```

subplot

first subplot-orignal price

```python
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o' ,label='Original price')
plt.plot(days, predicted_price, 'r-o',label='Predicted price')
plt.title('Original price over time')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend();


price_difference=[o-p for o,p in zip(original_price,predicted_price)]
plt.subplot(3,1,2)
plt.plot(days, price_difference, color='purple' ,label='Price difference')
plt.title('Difference between original and predicted')
plt.xlabel('Weekdays')
plt.ylabel('Price')
plt.legend();
plt.savefig('example_predicted.png',dpi=300)
```
